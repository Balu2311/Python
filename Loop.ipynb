```python
x=int(input("enter the number of rowa"))
y=2*x-1
n=y
space=y-n//2
while n>=1:
    for j in range(1,space+1):
        print(" ",end="")
    for i in range(1,n+1):
        print("*",end="")
    for k in range(1, space+1):
        print("",end="")
        n*n-2
```
